```python
raw_data1 = pd.read_csv('/kaggle/input/rotten-tomatoes-best-musicals-of-all-time/user_reviews.csv/user_reviews.csv')
raw_data1.head()
```

```python
data1 = raw_data1.drop(['userId','userRealm','userDisplayName','creationDate','reviewId','isSuperReviewer','hasProfanity'],axis=1)
data1.describe(include='all')
```

```python
raw_data2 = pd.read_csv('/kaggle/input/rotten-tomatoes-best-musicals-of-all-time/movies.csv/movies.csv')
raw_data2.head()
```

```python
data2 = raw_data2.drop(['movieURL'],axis=1)
data2.describe(include='all')
```

```python
data2.isnull().sum()
```

```python
#Merging data1 and data2 on movie id
data = pd.merge(data1,data2,on = 'movieId')
data.head()
```

```python
#drop the movieid column as it is not needed in further analysis
data_n = data.drop(['movieId','isVerified','hasSpoilers','movieYear','critic_score','audience_score'],axis=1)
data_n.head()
```

```python
#The quote is turned to sentiment and polarity score using textblob module
data_s = pd.DataFrame(data_n,columns=['quote'])

def analyze_sentiment(quote):
    blob = TextBlob(quote)
    return blob.sentiment.polarity, blob.sentiment.subjectivity

data_s['polarity'],data_s['subjectivity'] = zip(*data_s['quote'].apply(analyze_sentiment))
data_s.head()
```

```python
data_v = pd.merge(data_n,data_s,on = 'quote')
data_v.head()
```

```python
data_a = data_v.drop(['quote'],axis=1)
data_a.describe()
```

```python
#Sentiment of different movies based on user review
def classify_sentiment(polarity):
    if polarity>0:
        return 'positive'
    elif polarity<0:
        return 'negative'
    else:
        return 'neutral'

data_a['sentiment'] = data_a['polarity'].apply(classify_sentiment)
data_a
```
